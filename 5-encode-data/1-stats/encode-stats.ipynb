```python
import re

df = pd.read_csv('in.csv')


def encodeGender(row):
    text = str(row['Gender']).lower().strip()

    if "trans" in text or "transgender" in text:
        return 'T'
    elif "non-binary" in text or "nonbinary" in text or "non binary" in text:
        return 'NB'
    elif "female" in text or "girl" in text or text=='f' or "woman" in text:
        return 'F'
    elif "male" in text or "dude" in text or text=='m' or "man" in text or "guy" in text:
        #male INCLDUES Female SO MUST go after
        return 'M'
    elif "-1" in text:
        return 'MISSING'
    else:
        return 'OTHER'



def encodeRace(row):
    text = str(row['Race']).lower().strip()

    if "asian" in text or "indian" in text or "chinese" in text or "japanese" in text or "korean" in text or "filipino" in text or "vietnamese" in text or "thai" in text or "taiwanese" in text or "rice" in text or "pakistani" in text or "sri lank" in text or "pacific islander" in text or "hindu" in text or "korea" in text or "brown" in text or "tunak" in text or "Malaysi" in text:
        return "ASIAN"
    elif "white" in text or "caucasian" in text or "european" in text or "anglo" in text or "irish" in text or "italian" in text or "german" in text:
        return "WHITE"
    elif "black" in text or "african" in text or "haitian" in text or "jamaican" in text or "nigerian" in text or "ethiopian" in text or "somalian" in text or "kenya" in text:
        return "BLACK"
    elif "hispanic" in text or "latino" in text or "mexican" in text or "puerto rican" in text or "cuban" in text or "dominican" in text or "salvadoran" in text or "carribean" in text or "latino" in text or "latina" in text:
        return "HISPANIC"
    elif "arab" in text or "middle" in text or "persian" in text or "turkish" in text or "egyptian" in text or "syrian" in text or "iraqi" in text or "lebanese" in text or "jordanian" in text or "palestinian" in text or "israeli" in text or "yemeni" in text or "saudi" in text or "kuwaiti" in text or "qatari" in text or "emirati" in text or "omanian" in text or "bahraini" in text:
        return "ARAB"
    elif "native" in text:
        return "NATIVE"
    elif "mixed" in text or "multi" in text or "bi" in text:
        return "MIXED"
    elif "-1" in text:
        return "MISSING"
    else:
         return 'OTHER'
    
def encodeResidence(row):
    #1037

    text = str(row['Residence']).lower().strip()

    if "cali" in text or "bay" in text or "ca" in text or "yay area" in text or "los angeles" in text:
        return "CA"
    if "tx" in text or "texas" in text or "dallas" in text or "houston" in text or "austin" in text:
        return "TX"
    if "ny" in text or "new york" in text:
        return "NY"
    if "fl" in text or "florida" in text or "miami" in text or "orlando" in text or "tampa" in text:
        return "FL"
    if "il" in text or "chicago" in text:
        return "IL"
    if "seattle" in text or "wa" in text or "washington" in text or "pnw" in text:
        return "WA"
    if "colorado" in text or "denver" in text or "co" in text:
        return "CO"
    if "georgia" in text or "atlanta" in text or "ga" in text:
        return "GA"
    if "ohio" in text or "columbus" in text or "oh" in text:
        return "OH"
    if "michigan" in text or "detroit" in text or "mi" in text:
        return "MI"
    if "mass" in text or "boston" in text or "ma" in text:
        return "MA"
    if "dc" in text or "washington dc" in text or "dmv" in text or "dmv area" in text:
        return "DC"
    if "oregon" in text or "portland" in text or "or" in text:
        return "OR"
    if "arizona" in text or "phoenix" in text or "az" in text:
        return "AZ"
    if "minnesota" in text or "minneapolis" in text or "mn" in text:
        return "MN"
    if "tennessee" in text or "nashville" in text or "tn" in text:
        return "TN"
    if "pennsylvania" in text or "philly" in text or "pa" in text or "pittsburgh suburbs" in text:
        return "PA"
    if "wisconsin" in text or "milwaukee" in text or "wi" in text:
        return "WI"
    if "missouri" in text or "st. louis" in text or "mo" in text:
        return "MO"
    if "north carolina" in text or "charlotte" in text or "nc" in text:
        return "NC"
    if "nevada" in text or "las vegas" in text or "nv" in text:
        return "NV"
    if "indiana" in text or "indianapolis" in text or "in" in text:
        return "IN"
    if "maryland" in text or "baltimore" in text or "md" in text:
        return "MD"
    if "virginia" in text or "richmond" in text or "va" in text:
        return "VA"
    if "utah" in text or "salt lake" in text or "ut" in text:
        return "UT"
    if "oklahoma" in text or "okc" in text or "ok" in text:
        return "OK"
    if "kansas" in text or "kansas city" in text or "ks" in text or "kanas" in text:
        return "KS"
    if "louisiana" in text or "new orleans" in text or "la" in text:
        return "LA"
    if "alabama" in text or "birmingham" in text or "al" in text:
        return "AL"
    if "iowa" in text or "des moines" in text or "ia" in text:
        return "IA"
    if "idaho" in text or "boise" in text or "id" in text:
        return "ID"
    if "hawaii" in text or "honolulu" in text or "hi" in text:
        return "HI"
    if "alaska" in text or "anchorage" in text or "ak" in text:
        return "AK"
    if "montana" in text or "mt" in text:
        return "MT"
    if "wyoming" in text or "wy" in text:
        return "WY"
    if "south dakota" in text or "sd" in text:
        return "SD"
    if "north dakota" in text or "nd" in text:
        return "ND"
    if "nebraska" in text or "ne" in text:
        return "NE"
    if "new mexico" in text or "nm" in text:
        return "NM"
    

    if "midwest" in text:
        return "MIDWEST"
    if "suburban" in text or "suburb" in text or "suburbs" in text or "suburban town" in text or "suburban nj" in text or "suburban nj town" in text or "suburban (not a very big state)" in text:
        return "SUBURBAN"
    if any(country in text for country in ["brazil", "india", "asia", "china", "turkey", "puerto", "europe", "dubai", "uk", "abu dhabi, uae", "guam, u.s.", "taipei", "accra,ghana", "ghana", "uae", "sweden", "hk", "qatar", "uzbekistan", "hong kong", "sg"]):
        return "INTERNATIONAL"
    if "-1" in text:
        return "MISSING"
    if "u.s." in text or "united states" in text or "usa" in text or "us citizen" in text or "uh…usa? the west" in text:
        return "USA"

    return "OTHER"



def encodeIncome(row):
    text = str(row['Income']).lower().strip()
    number = re.sub("[^0-9]", "", text)

    if number.isnumeric():
        number = int(number)
        if number < 50000:
            return "Low"
        elif 50000 <= number <= 150000:
            return "Middle"
        else:
            return "High"

    if any(keyword in text for keyword in ["middle", "200", "250", "high", "full pay", "upper", "no aid", "no financial aid", "did not apply", "full-pay", "didn't apply", "300,000", "rich", "wealthy", "won't qualify", "did not qualify", "very well off", "comfortable", "don’t qualify for aid", "did not ask for aid", "does not qualify", "doesn't qualify"]):
        return "High"

    if any(keyword in text for keyword in ["low", "poor", "30,000", "60,000", "broke", "pell eligible", "pell grant", "free lunch", "need a lot of aid", "negative sai" ]):
        return "Low"

    if "-1" in text or "nan" in text or "n/a" in text or "prefer to not disclose" in text or "undisclosed" in text or "not sure" in text or "no idea" in text or "idk" in text or "unknown" in text:
        return "MISSING"

    return "OTHER"



def encodeSchool(row):
    text = row['School Type'].lower().strip()

    if "public" in text:
        return "PUBLIC"
    if "private" in text:
        return "PRIVATE"
    if "charter" in text:
        return "CHARTER"
    
    if "small" in text or "rural" in text or "tiny" in text:
        return "SMALL"
    if "international" in text:
        return "INTERNATIONAL"
    if "large" in text:
        return "LARGE"
    if "magnet" in text:
        return "MAGNET"
    
    if "-1" in text:
        return "MISSING"

    return "OTHER"

def encodeHooks(row):
    text = str(row['Hooks']).lower().strip()

    if 'urm' in text:
        return "URM"
    if 'first gen' in text or 'first generation' in text or 'first-gen' in text:
        return "FIRST-GEN"
    
    if 'legacy' in text:
        return 'LEGACY'
    
    if 'gay' in text or 'trans' in text or 'lgbt' in text:
        return "LGBT"
    
    if any(keyword in text for keyword in ["nan", "na", "none", "nada", "nothing", "nope"]):
        return "NONE"

    if "-1" in text:
        return "MISSING"
    
    return "OTHER"



def encodeMajor(row):
    text = str(row['Major']).lower().strip()

    if any(keyword in text for keyword in ["cs", 'computer science', 'comp sci']):
        return "CS"
    
    if any(keyword in text for keyword in ["engineering", 'eng']):    
        return "ENG"
    
    if any(keyword in text for keyword in ["business", 'finance', 'economics', 'accounting', 'marketing', 'management', 'mba', 'business administration']):
        return "BUSINESS"
    
    if any(keyword in text for keyword in ["biology", 'bio', 'biomedical', 'biotech', 'biological']):
        return "BIO"
    
    if any(keyword in text for keyword in ["psychology", 'psych', 'counseling', 'therapy']):
        return "PSYCH"
    
    if any(keyword in text for keyword in ["premed", 'pre-med', 'med', 'medical', 'pre med', 'premedical']):
        return "PREMED"
    
    if any(keyword in text for keyword in ["prelaw", 'pre-law', 'law', 'legal', 'lawyer', 'attorney']):
        return "LAW"
    
    if any(keyword in text for keyword in ['poli sci', 'polisci', 'political science', 'political']):
        return 'POLISCI'

    if any(keyword in text for keyword in ['nan', '(write here)', '-1']):
        return "MISSING"

    if any(keyword in text for keyword in ["neuroscience", "cognitive science", "behavior"]):
        return "NEUROSCIENCE"

    if any(keyword in text for keyword in ["sociology", "social work", "global studies", "anthropology"]):
        return "SOCIAL SCIENCES"

    if any(keyword in text for keyword in ["history", "art history", "history of art", "liberal studies", "humanities"]):
        return "HISTORY"

    if any(keyword in text for keyword in ["english", "literature", "creative writing", "comparative literature", "philosophy", "classics", "languages", "linguistics"]):
        return "HUMANITIES"

    if any(keyword in text for keyword in ["environmental", "ecology", "geography", "earth sciences", "environmental health science", "environmental science/studies"]):
        return "ENVIRONMENTAL STUDIES"

    if any(keyword in text for keyword in ["chemistry", "chemical engineering", "biochemistry", "pharmaceutical sciences", "pharmacology"]):
        return "CHEMISTRY"

    if any(keyword in text for keyword in ["physics", "astronomy", "astrophysics"]):
        return "PHYSICS"

    if any(keyword in text for keyword in ["math", "mathematics", "applied math", "statistics", "data science"]):
        return "MATH"

    if any(keyword in text for keyword in ["public health", "health science", "nursing", "health promotion", "health policy"]):
        return "HEALTH SCIENCES"

    if any(keyword in text for keyword in ["international relations", "global studies", "foreign service", "diplomacy", "international studies"]):
        return "INTERNATIONAL RELATIONS"

    if any(keyword in text for keyword in ["film", "media", "communications", "journalism", "broadcasting", "tv production", "film/tv production", "animation"]):
        return "MEDIA STUDIES"

    if any(keyword in text for keyword in ["design", "graphic design", "communications design", "industrial design", "interior design", "architecture", "urban planning"]):
        return "DESIGN"

    if any(keyword in text for keyword in ["music", "jazz studies", "music education", "music industry", "music theory", "music technology", "composition", "performance", "classical composition", "musical theatre"]):
        return "MUSIC"

    if any(keyword in text for keyword in ["education", "secondary education", "teaching"]):
        return "EDUCATION"

    if any(keyword in text for keyword in ["theatre", "drama", "acting", "production and design for stage/screen"]):
        return "THEATRE"

    if any(keyword in text for keyword in ["arts", "fine arts", "studio art", "visual arts", "illustration", "photography"]):
        return "ARTS"

    if any(keyword in text for keyword in ["urban studies", "urban planning", "city planning"]):
        return "URBAN STUDIES"

    if any(keyword in text for keyword in ["public policy", "policy", "government", "politics"]):
        return "PUBLIC POLICY"

    if any(keyword in text for keyword in ["preprofessional programs", "professional programs", "undecided", "undeclared", "liberal arts", "general studies"]):
        return "UNDECIDED"

    return "OTHER"


def encodeRank(row):
    text = str(row['Rank']).lower().strip()


    if "%" in text:
        pattern = re.compile(r"((\d)*%)", re.IGNORECASE)
        match = pattern.search(text)
        return match.group(1)
    
    if "/" in text and not '.' in text:
        #(\d*)\/(\d*)
        pattern = re.compile(r"(\d*)\/(\d*)", re.IGNORECASE)
        match = pattern.search(text)

        #1/450
        num = 0
        if match.group(1).isnumeric():
            num= int(match.group(1))

        
        denom = 0
        if match.group(2).isnumeric():
            denom = int(match.group(2))


        if not (num ==0 or denom == 0):
            divided = num/denom
            divided *= 100
            divided = round(divided)


            if divided == 0:
                divided = 1

            if not divided > 99:
                return str((divided)) + "%"
        
        #print(num, denom, text)


    if "-1" in text or "none" in text or "n/a" in text or "doesn't rank" in text or "doesn’t rank" in text or "no rank" in text or "does not rank" in text or "doesn't" in text or "don't" in text or "not ranked" in text:
        return "MISSING"


    return "OTHER"



def encodeSAT(row):
    text = str(row['SAT']).lower().strip()
    #(\d\d\d\d)
    pattern = re.compile(r"(\d\d\d\d)", re.IGNORECASE)
    match = pattern.search(text)
    if match:
        #print(match.group(1))
        return (match.group(1))

    if "didn't take" in text or "did not submit" in text or "didn't" in text or "did not" in text or "-1" in text or "error" in text:
        return "MISSING"
    
    return "OTHER"


def encodeACT(row):
    text = str(row['ACT']).lower().strip()
    #(\d\d)
    pattern = re.compile(r"(\d\d)", re.IGNORECASE)
    match = pattern.search(text)
    if match:
        #print(match.group(1))
        if int(match.group(1)) <= 36:
            return (match.group(1))
    
    if "didn't take" in text or "did not submit" in text or "didn't" in text or "did not" in text or "-1" in text or "error" in text:
        return "MISSING"
    
    return "OTHER"



df["Gender"] = df.apply(encodeGender, axis=1)
df["Race"] = df.apply(encodeRace, axis=1)
df['Residence'] = df.apply(encodeResidence, axis=1)
df["Income"] = df.apply(encodeIncome, axis=1)
df['School Type'] = df.apply(encodeSchool, axis=1)
df['Hooks'] = df.apply(encodeHooks, axis=1)
df['Major'] = df.apply(encodeMajor, axis=1)
df['Rank'] = df.apply(encodeRank, axis=1)
df['SAT'] = df.apply(encodeSAT, axis=1)
df['ACT'] = df.apply(encodeACT, axis=1)



df.drop(columns=['Body Text', 'APs', 'GPA', 'SeniorCourseLoad'], inplace=True)


df.to_csv("out.csv", index=False)
```
